```python
#EJERCICIO 1 OBLIGATORIO ENOL

MAX_ITERACIONES = 1000000

def init():
    print("[1] Metodo de la biseccion.")
    print("[2] Metodo de la falsa posicion.")
    print("[3] Salir")
    respuesta = int(input("Respuesta: "))
    if(respuesta == 1):
        biseccion()
    if(respuesta == 2):
        iteraciones = 0
        error, limiteInferior, limiteSuperior = cargarDatos()
        regulaFalsi(error, limiteInferior, limiteSuperior, iteraciones)
    if(respuesta == 3):
        return print("Adios.")
    else:
        print("Introduzca 1 o 2 para elegir el metodo.")
        init()

def cargarDatos():
    error = float(input("Ingrese el error: "))
    limiteInferior = float(input("Ingrese el limite inferior: "))
    limiteSuperior = float(input("Ingrese el limite superior: "))
    return error, limiteInferior, limiteSuperior

def f(x):
    y = x**(4)-2*x**(3)-4*x**(2)+4*x-4 #fixed function
    return y

def obtenerPuntoBiseccion(limiteInferior, limiteSuperior):
    puntoMedio = (limiteSuperior + limiteInferior)/2
    return puntoMedio

def obtenerPuntoRegulaFalsi(limiteInferior, limiteSuperior):
    nuevoPunto = (limiteInferior*f(limiteSuperior) - limiteSuperior*f(limiteInferior))/(f(limiteSuperior)-f(limiteInferior))
    return nuevoPunto

def regulaFalsi(error, limiteInferior, limiteSuperior, iteraciones):
    if(iteraciones > MAX_ITERACIONES):
        return print("Maximo de iteraciones alcanzado.")
    if(f(limiteInferior)*f(limiteSuperior) > 0):
        return print("ERROR EN EL INTERVALO DE ENTRADA")
    puntoMedio = obtenerPuntoRegulaFalsi(limiteInferior,limiteSuperior)
    if abs(f(puntoMedio)) > error:
        iteraciones = iteraciones + 1
        if(f(limiteInferior)*f(puntoMedio) < 0):
            return regulaFalsi(error,limiteInferior,puntoMedio, iteraciones)
        if(f(limiteSuperior)*f(puntoMedio) < 0):
            return regulaFalsi(error,puntoMedio,limiteSuperior, iteraciones)
    if(f(puntoMedio) < error):
        print("Iteraciones:",iteraciones)
        return print("Raiz: ",puntoMedio)
    

def biseccion():
    error, limiteInferior, limiteSuperior = cargarDatos()
    if((f(limiteInferior) * f(limiteSuperior)) > 0):
        return print("ERROR EN EL INTERVALO DE ENTRADA")
    iteraciones = 0
    puntoMedio = obtenerPuntoBiseccion(limiteInferior, limiteSuperior)
    while abs(f(puntoMedio)) > error:
        iteraciones = iteraciones + 1
        if(f(limiteInferior)*f(puntoMedio) > 0):
            limiteInferior = puntoMedio
        else:
            limiteSuperior = puntoMedio
        puntoMedio = obtenerPuntoBiseccion(limiteInferior, limiteSuperior)
        if(iteraciones > MAX_ITERACIONES):
            return print("Maximo de iteraciones alcanzado.")
    print("Iteraciones: ",iteraciones)
    return print("Raiz: ",puntoMedio)
    
init()
```
